```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import pandas as pd
import numpy as np
import random
from copy import deepcopy
```

```python
df = pd.read_csv('grad.csv')
df = df.iloc[0:25]
df = df[['GRE Score','Chance of Admit ']]
```

```python
fig, [ax1, ax2] = plt.subplots(1, 2, figsize=(12,4))
x = df['GRE Score'] / 100
y = df['Chance of Admit '] * 10
points = np.array(list(zip(x, y)))
ax1.scatter(x, y)
ax1.set_xlabel('Independent')
ax1.set_ylabel('Dependent')
ax2.set_xlabel('Iterations')
ax2.set_ylabel('Loss')
```

```python
SLOPE_LEARNING_RATE = .005
INTERCEPT_LEARNING_RATE = .0001
BATCH_SIZE = 5

def SS(m, b):
    ss = 0
    for point in points:  
        ss += ((point[1]-(m*point[0]+b))**2)
    return ss

def select_batch(data):
    batch = []
    for i in range(BATCH_SIZE):
        batch.append(points[random.randint(0, 24)])
    return batch

def intercept_gradient(m, b, data):
    i_grad = 0 
    for d in data:  
        i_grad += (-2)*(d[1]-(m*d[0] + b)) 
    return i_grad  

def slope_gradient(m, b, data):
    s_grad = 0
    for d in data:
        s_grad += (-2 * d[0])*(d[1]-(m*d[0] + b))
    return s_grad

def new_intercept(m, b, data):
    intercept_step_size = INTERCEPT_LEARNING_RATE * intercept_gradient(m, b, data)
    b = b - intercept_step_size
    return b

def new_slope(m, b, data):
    slope_step_size = SLOPE_LEARNING_RATE * slope_gradient(m, b, data)
    m = m - slope_step_size
    return m
```

```python
points2 = []
slope = 0
intercept = 0
epoch = 0
x1 = np.linspace(2.95, 3.4)

def animate():
    global points2
    global slope
    global intercept
    global epoch
    y1 = slope*x1 + intercept
    ax1.plot(x1, y1, label='My model')
    
    y2 = SS(slope, intercept)
    
    batch = select_batch(points)
    inter = deepcopy(intercept)
    intercept = new_intercept(slope,intercept,batch)
    slope = new_slope(slope,inter,batch)
    
    points2.append([epoch, y2])
    
    if len(ax2.lines) > 0:
        ax2.lines[0].remove()
    ax2.scatter(epoch, y2, s=20)
    ax2.plot(np.array(points2)[:, 0], np.array(points2)[:, 1], label='My model', color='darkturquoise')
    epoch += 1
    if len(ax1.lines) > 1:
        ax1.lines[0].remove()
```

```python
for i in range(40):
    animate()
ax1.set_ylim(2, 10)
ax2.set_ylim(20, 70)
ax1.set_title('Regression using Stochastic Gradient Descent')
ax2.set_title(f'Data Size: {len(points)}  Batch Size: {BATCH_SIZE}')
fig
```

# Changing the starting Intercept

```python
points3 = []
slope3 = 0
intercept3 = -30  
epoch3 = 0
x3 = np.linspace(2.95, 3.4)

def animate():
    global points3
    global slope3
    global intercept3
    global epoch3
    y3 = slope3*x1 + intercept3
    ax1.plot(x3, y3, label='My model 2.0', color='goldenrod')
    
    y3 = SS(slope3, intercept3)
    
    batch = select_batch(points)
    inter = deepcopy(intercept3)
    intercept3 = new_intercept(slope3,intercept3,batch)
    slope3 = new_slope(slope3,inter,batch)
    
    points3.append([epoch3, y3])
    
    if len(ax2.lines) > 1:
        ax2.lines[1].remove() 
    ax2.scatter(epoch3, y3, s=20)
    ax2.plot(np.array(points3)[:, 0], np.array(points3)[:, 1], label='My model 2.0', color='goldenrod')
    
    epoch3 += 1
    if len(ax1.lines) > 3:
        ax1.lines[2].remove()
```

```python
for i in range(40):
    animate()
ax1.set_ylim(2, 10)
ax2.set_ylim(20, 70)
ax1.legend()
ax2.legend()
fig
```
